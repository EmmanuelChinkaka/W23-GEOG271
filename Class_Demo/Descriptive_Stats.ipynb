```python
# Import packages, This is how we import the packages 

import pandas as pd
```

```python
# Import data using URL

data = pd.read_csv("https://raw.githubusercontent.com/kirenz/datasets/master/wage.csv")
```

```python
# Summary of the stats of the numerial variables 
data.describe()
```

```python
# Accessing specific variables 

wage = data['wage']
wage
```

```python
# Calculating the the mean wage
mean_wage = round(wage.mean(), 4)
print('The mean wage in the data set is ', mean_wage)

mode_wage = round(wage.mode(), 4)
print('The mode wage in the data set is ', mode_wage)

median_wage = round(wage.median(), 4)
print('The median wage in the data set is ', median_wage)
```

```python
# Max, min, Range


print('Max age is ', data['age'].max())

print('Min age is ', data['age'].min())

print('Range age is ', data['age'].max() - data['age'].min())
```

```python
# std

print('Std of variable age is ', round(data['age'].std(), 4))
```

```python
# Stats summary by variables

x = data['age'].groupby(data['education'])
x.describe()
```

```python
mean = round(data['age'].mean(), 2)
print('Mean is ', mean)

median = round(data['age'].median(), 2)
print('Median is ', median)
```

```python
# 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Histogram

ax = data['age'].hist(bins=20)
ax.set_xlabel('Age')
ax.set_ylabel('Frequency')
ax.set_title('Age Distribution')
ax.axvline(mean, color='red', linestyle='solid', linewidth=2, alpha= .8)
ax.axvline(median, color='green', linestyle='solid', linewidth=2, alpha= .8)
ax.legend([mean, median], labels = ['Mean', 'Median']);


```

```python
nu_cols = ['age', 'logwage', 'wage']
nu_cols
```

```python
sns.pairplot(data[nu_cols], size=2.0)
```

```python
sns.distplot(data['age'], bins=20)
plt.title('Probability Distribution')
plt.show()
```
